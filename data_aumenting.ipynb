```python
import json
import pandas as pd
 
# Opening JSON file
f = open('openai.json')
 
# returns JSON object as
# a dictionary
data = json.load(f)
import os
import openai
openai.api_key =data['key']

prompt = '''

Create 10 similar reviews, please try to use similar but different words (start each review with a - and just one line between samples)

Current Review: {}

New paraphrased reviews:


'''

manual_syntethic  = pd.read_csv('Syntetic_reviews/reviews_all.csv')
```

```python
news= []

df = manual_syntethic
for index, row in df.iterrows():
    print('{} out of {}'.format(index,df.shape[0]))
 
    aux = openai.Completion.create(
      model="text-davinci-003",
      prompt=prompt.format(row['Review']),
      max_tokens=200,
      temperature=1
    )
    new_reviews = [s[1:]  for s in aux.to_dict_recursive()['choices'][0]['text'].split('\n') if s!= '']
    new = pd.DataFrame(new_reviews)
    new.columns = ['Review']
    new['Topic'] = row['Topic']
    new['original'] = row['Review']
    news.append(new)
news = pd.concat(news)

all = pd.concat([df[['Review','Topic']],news[['Review','Topic']]])
```
