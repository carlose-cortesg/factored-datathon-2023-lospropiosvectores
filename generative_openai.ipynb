```python
import json
import pandas as pd
 
# Opening JSON file
f = open('openai.json')
 
# returns JSON object as
# a dictionary
data = json.load(f)
import os
import openai
openai.api_key =data['key']


def generate_review(topic):
    prompt = 'write a small review (one sentence) talking about a shirt, a bag or a watch talking about {} (it can be good or bad):'.format(topic)
    aux = openai.Completion.create(
      model="text-davinci-003",
      prompt=prompt,
      max_tokens=200,
      temperature=1
    )
    return aux.to_dict_recursive()['choices'][0]['text'].replace('\n',' ').strip().replace('"','')

topics = list(pd.read_csv('Syntetic_reviews/reviews_all.csv').Topic.unique())

reviews = []
for i in range(100):
    print(i)
    for topic in topics:
        new = generate_review(topic)
        reviews.append((topic,new))

pd.DataFrame(reviews, columns = ['Topic','Review']).to_csv('Syntetic_reviews/openai_syn.csv')

```
